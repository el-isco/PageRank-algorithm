Notice that earlier I subtracted 1 to each node(page) number while making the dataframe. If I hadn't done that, the sparse P would have to have shape of (685231, 685231). This is because the enumeration of the pages starts from 1, but python matrices dimensions start from dimension zero, thus an empty row 0 and a column 0 would also have been included in the sparse matrix.

**So, now the sparse matrix is of dimensions 685230x685230 and page 0 is actually initial page 1, page 1 is actually initial page 2 and so on.**

```python
Psparse = csr_matrix((probabilities_flat_list, (np.asarray(data.Starting_node), np.asarray(data.Ending_node))), shape=(max(data.Starting_node)+1, max(data.Starting_node)+1)) # max+1=685230
```

```python
def power_method(a, tol, dimension):
  # a = 0.85
  # tol = 10**(-8)                                 # tolerance
  alpha = np.ones(max(dimension)+1)                # max+1=685230
  for i in unique_elements:
    alpha[i] = 0
  n = max(dimension)+1
  v = np.repeat(1/n, n)
  previous_x = np.repeat(1/n, n)

  iter = 0
  condition = 1
  while condition>tol:
    new_x = a*previous_x.T*Psparse + (a*previous_x.T.dot(alpha) + (1-a))*v.T
    condition = np.linalg.norm(new_x-previous_x)
    iter += 1
    previous_x = new_x
  
  return sum(previous_x), iter, new_x     # sum of normalized "pageranks", converges in __ iterations, pageranks' vector
```

**Now, I am going to define a new function for the power_method but also with max(iterations) as input, in order to timesnap the convergence of the values of vector π.**

```python
def power_method_with_max_iter(a, tol, dimension, max_iterations):

  alpha = np.ones(max(dimension)+1)                # max+1=685230
  for i in unique_elements:
    alpha[i] = 0
  n = max(dimension)+1
  v = np.repeat(1/n, n)
  previous_x = np.repeat(1/n, n)

  iter = 0
  condition = 1
  while condition>tol and iter<max_iterations:                         # !!!!! Here is the only change added to the function
    new_x = a*previous_x.T*Psparse + (a*previous_x.T.dot(alpha) + (1-a))*v.T
    condition = np.linalg.norm(new_x-previous_x)
    iter += 1
    previous_x = new_x
  
  return sum(previous_x), iter, new_x              # sum of normalized "pageranks", converges in __ iterations, pageranks' vector
```

Remember that with α=0.85 and tol=10e-8, 71 iterations take place before convergence.

```python
iter_snaps = [15, 30, 45, 50, 65, 70, 71]
pi_vectors_list = []                               # here, the π vectors will be added of 15,30,45,60,70,71 iterations
for i in iter_snaps:
  pi_vectors_list.append(power_method_with_max_iter(0.85, 10**(-8), data.Starting_node, i)[2])
```

Redefining the power_method adding to it the P_sparse matrix as input:

```python
def power_method_P_sparse(a, tol, dimension, P_sparse):

  alpha = np.ones(max(dimension)+1)                # max+1=685231
  for i in unique_elements:
    alpha[i] = 0
  n = max(dimension)+1
  v = np.repeat(1/n, n)
  previous_x = np.repeat(1/n, n)

  iter = 0
  condition = 1
  while condition>tol:
    new_x = a*previous_x.T*P_sparse + (a*previous_x.T.dot(alpha) + (1-a))*v.T
    condition = np.linalg.norm(new_x-previous_x)
    iter += 1
    previous_x = new_x
  
  return sum(previous_x), iter, new_x     # sum of normalized "pageranks", converges in __ iterations, pageranks' vector
```

```python
question_4 = power_method_P_sparse(0.85, 10**(-8), data_1.Starting_node, Psparse4)      # ~3 seconds
print("Using the power method the new node, 685230, got the rank:", len(question_4[2])-stats.rankdata(question_4[2], method='ordinal')[-1] + 1)
```
